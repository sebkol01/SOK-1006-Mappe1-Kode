```python
df=pd.read_csv("https://uit-sok-1006-v22.github.io/innleveringer/chickendata.csv",delimiter=";")
```

```python
chicken.plot_year(df,1995)
```

### Beskrivelse av figuren
Figuren over gjelder for året 1995. På figuren over ser vi antall enheter kylling på x-aksen og pris per enhet på y-aksen. Den grønne linjen viser etterspørselskurven, den blå viser den kortsiktige tilbudskurven og den røde viser den langsiktige tilbudskurven.

```python
chicken.plot_year(df, 1960)
```

### Beskrivelse av Figur
Figuren over gjelder for året 1960. På figuren over ser vi antall enheter kylling på x-aksen og pris per enhet på y-aksen. Det er en linje som viser etterspørselskurven, en som viser den kortsiktige tilbudskurven og en som viser den langsiktige tilbudskurven.

### Kommentar på grafene
Skiftet i etterspørsel fra 1960 til 1995 skyldes en økning i inntekt per capita på omkring 225% over denne perioden. Prisen på substituttgodet, biff, har gått ned med omkring 25% over perioden. Selv om denne nedgangen i pris for substitutten forskjøv en del av vekst for etterspørsel av kylling, er denne effekten lav ifht. veksten i inntekt per capita. 

Skiftet utover i tilbudskurven er et resultat av et fall i prisen på kyllingfor på omtrendt 50% og en betydelig økning i produktivitet av kyllingproduksjon. Sistnevnte er fanget opp av koeffisienten 0,0183 på tidsvariabelen i tilbudsfunksjonen. Vi kan se at forskyvningen var mer raskere for tilbudskurven enn det var for etterspørselskurven som førte til et betydelig fall i prisen på kylling over intervallet på 35 år.
